```python
import tensorflow
from keras.preprocessing import image
from keras.layers import  GlobalMaxPooling2D
from keras.applications.resnet50 import ResNet50
import numpy as np
from keras.applications.resnet50 import preprocess_input
```

```python
model=ResNet50(weights="imagenet",include_top=False,input_shape=(224,224,3))
model.trainable=False
```

```python
model=tensorflow.keras.Sequential([
    model,
    GlobalMaxPooling2D()

])
```

```python
def extract_feature(img_path,model):
  img=image.load_img(img_path,target_size=(224,224))
  img_array=image.img_to_array(img)
  expand_img_array=np.expand_dims(img_array,axis=0)
  preprocessed_img=preprocess_input(expand_img_array)
  result=model.predict(preprocessed_img).flatten()
  normalized_result=result/norm(result)

  return normalized_result


```

```python
import os
```

```python
filenames=[]
for file in os.listdir("/content/myntradataset/images"):
  filenames.append(os.path.join("images",file))

```

```python
len(filenames)
```

```python
print(filenames[0:5])
```

```python
feature_list=[]
for file in filenames:
  feature_list.append(extract_feature(file,model))

```

```python
print(np.array(feature_list).shape)
```

```python
import pickle
pickle.dump(feature_list,open("embedding.pkl","wb"))
pickle.dump(filenames,open("filenames.pkl","wb"))
```
